################   爱好与生活习惯数据集 ################

```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv('./dataset/关联分析/爱好与生活习惯调查数据集/responses.csv')
df.head()
```

```python
data=df.iloc[:,:10]
data
```

```python
columns=data.columns
data
```

```python
data=pd.DataFrame(np.where(data>=3, int(1), int(0)),columns=columns)
data
```

```python
# 计算频繁项集
frequent_itemsets = apriori(data, min_support=0.7, use_colnames=True)
print(frequent_itemsets)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules


# 根据频繁项集生成关联规则
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
print(rules)
```

```python
# 打印关联规则
for _, rule in rules.iterrows():
    antecedents = ", ".join([str(item) for item in rule['antecedents']])
    consequents = ", ".join([str(item) for item in rule['consequents']])
    print(f"Rule: {antecedents} -> {consequents}")
    print(f"Support: {rule['support']}")
    print(f"Confidence: {rule['confidence']}")
    print()
```

```python
import os
import pandas as pd
import numpy as np
import sys
from operator import itemgetter
from collections import defaultdict
```

```python
# 加载数据
all_ratings = pd.read_csv("./dataset/关联分析/电影推荐数据集/ratings.csv")
all_movies =pd.read_csv('./dataset/关联分析/电影推荐数据集/movies.csv')
```

```python
# 使用 merge 方法连接两个 DataFrame
merged_df = pd.merge(all_ratings,all_movies, on='movieId')
merged_df
```

```python
# 电影分数都有哪几档
merged_df["rating"].unique().tolist()
```

```python
# 解析时间戳
merged_df["timestamp"] = pd.to_datetime(merged_df["timestamp"], unit='s')
merged_df["timestamp"]
```

```python
# 用户是不是喜欢某一部电影，评分高于3的认定为喜欢
merged_df['Favorable'] = merged_df['rating'] > 3
merged_df
```

```python
user_movie = merged_df.pivot_table(index = 'userId', columns = 'title', values = 'Favorable')
user_movie
```

```python
bool_data = user_movie.notnull()
bool_data
```

```python
# 计算频繁项集
frequent_itemsets = apriori(bool_data, min_support=0.3,use_colnames=True)
print(frequent_itemsets)
```

```python
# 根据频繁项集生成关联规则
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
print(rules)
```

```python
# 打印关联规则
for _, rule in rules.iterrows():
    antecedents = ", ".join([str(item) for item in rule['antecedents']])
    consequents = ", ".join([str(item) for item in rule['consequents']])
    print(f"Rule: {antecedents} -> {consequents}")
    print(f"Support: {rule['support']}")
    print(f"Confidence: {rule['confidence']}")
    print()
```

########################################## 商场购物数据集##############################################

```python
import pandas as pd

# 创建购买数据的列表
purchase_data = []

# 打开文本文件并按行读取
with open('./dataset/关联分析/商场购物记录数据集/shopping-record.txt', 'r') as file:
    # 逐行读取文件内容
    for line in file:
        # 移除行末尾的换行符和空格
        line = line.strip()
        # 检查行是否为空
        if line:
            # 将逗号分隔的商品列表拆分成一个商品数组
            items = line.split(',')
            # 将商品数组存储到购买数据列表中
            purchase_data.append(items)

# 获取所有唯一的商品列表
unique_items = set([item for purchase in purchase_data for item in purchase])

# 创建一个空的字典来存储是否购买商品的信息
purchase_dict = {item: [] for item in unique_items}

# 填充购买字典
for purchase in purchase_data:
    for item in unique_items:
        if item in purchase:
            purchase_dict[item].append(1)  # 商品被购买
        else:
            purchase_dict[item].append(0)  # 商品未被购买

# 创建是否购买商品的DataFrame
df = pd.DataFrame(purchase_dict)

# 打印DataFrame
for i in df.columns:
    print(df[i].value_counts())
```

```python
# 计算频繁项集
frequent_itemsets = apriori(df, min_support=0.01,use_colnames=True)
print(frequent_itemsets)
```

```python
# 根据频繁项集生成关联规则
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
print(rules)
```

```python
# 打印关联规则
for _, rule in rules.iterrows():
    antecedents = ", ".join([str(item) for item in rule['antecedents']])
    consequents = ", ".join([str(item) for item in rule['consequents']])
    print(f"Rule: {antecedents} -> {consequents}")
    print(f"Support: {rule['support']}")
    print(f"Confidence: {rule['confidence']}")
    print()
```
